```python
import numpy as np
from pydub import AudioSegment
import random
import sys
import io
import os
import glob
import IPython
from td_utils import *
%matplotlib inline
```

```python
def insert_audio_clip(background, audio_clip):    
    # Get the duration of the audio clip in ms
    segment_ms = len(audio_clip)
    print('org_segment_ms:', segment_ms)
    if segment_ms > 10000:
        segment_ms = 10000
    segment_time = (0, segment_ms)
    print('segment_time:', segment_time)
    
    # Superpose audio segment and background
    new_background = background.overlay(audio_clip, position = segment_time[0])
    
    return new_background
```

```python
 _, _, backgrounds = load_raw_audio_thalon_list()
print(backgrounds[0])
audio_org = './audio_examples/thalon/output.wav'
audio_list = []
if os.path.exists(audio_org):
    audio_list.append(audio_org)
audio = load_raw_data(audio_list)

background = load_raw_data(random.sample(backgrounds, k=1))
background = background[0] - 20

background = insert_audio_clip(background, audio[0])
background = match_target_amplitude(background, -20.0)
file_handle = background.export("./audio_examples/thalon/test_output.wav", format="wav")
x = graph_spectrogram("./audio_examples/thalon/test_output.wav")
```
